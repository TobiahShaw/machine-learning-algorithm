# 主成分分析（Principal Component Analysis）
- 不仅应用于机器学习，统计学常用
- 一个非监督的机器学习算法
- 主要用于数据降维
- 通过降维可以发现更便于人类理解的特征
- 其他应用：可视化；去噪

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = [
    [1, 2, 3, 4, 5],
    [1.3 , 1.5, 2.5, 4.8, 5.2]
]
```

```python
plt.scatter(X[0], X[1])
plt.axis([0, 6 ,0, 6])
plt.show()
```

## 降维

```python
plt.scatter(X[0], np.zeros(len(X[1])))
plt.axis([0, 6 ,0, 6])
plt.show()
```

```python
plt.scatter(np.zeros(len(X[0])), X[1])
plt.axis([0, 6 ,0, 6])
plt.show()
```

**以上两种为忽略其中某个特征的降维方法，即投射到两个特征轴的某一个轴**

```python
plt.scatter(X[0], X[1])
plt.plot(X[0], X[0])
plt.axis([0, 6 ,0, 6])
plt.show()
```

**也可以将样本点投射到上述特征轴上，此时，每个样本点在特征空间的距离比较大，我们可以寻找一条线，使得每个样本点在特征空间距离最大，从而与原样本点之间的距离差距更好，样本间的区分度更大，并且，即可以求方差的最大化来找到这条线**
